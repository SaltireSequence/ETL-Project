```python
# Dependencies and Setup
import pandas as pd
import geopy
from geopy.geocoders import Nominatim
import json
from pprint import pprint
from pymongo import MongoClient
import gmaps
# Import API key
from api_keys import g_key
```

```python
# File to Load 
file_to_load_pika_boba = "pika_boba.csv"
pika_boba_df = pd.read_csv(file_to_load_pika_boba)
pika_boba_df.head()
```

```python
#gmap was used to make a map
gmaps.configure(api_key=g_key)
```

```python
# boba shops were sorted with rating and their information was made into data frames.
pika_boba_df_sorting = pika_boba_df.sort_values("rating",ascending=False)
pika_boba_df_sorting
top_boba_df =pd.DataFrame(pika_boba_df_sorting)
top_boba_df
```

```python
#The areas where 100 Pikachu appear most frequently are filtered and their information is made into data frames.
pika_boba_df_sorting = pika_boba_df.sort_values("Number of Pikachu",ascending=False)
pika_boba_df_sorting.head(100)
top_100_pika_df =pd.DataFrame(pika_boba_df_sorting.head(100))
top_100_pika_df
```

```python
# 1000 of the best rating parks were filtered and their information was made into data frames.
file_to_load_parks = "Parks.csv"
parks_df = pd.read_csv(file_to_load_parks)
parks_df_sorting = parks_df.sort_values("Score",ascending=False)
parks_df_sorting.head(1000)
top_1000_parks_df =pd.DataFrame(parks_df_sorting.head(1000))
top_1000_parks_df
```

```python
# load and read the SF rastaurant file
file_to_load_resaurant = "Will_data.csv"
resaurant_df = pd.read_csv(file_to_load_resaurant)
resaurant_df
```

```python
#create info_box_template
info_box_template = """
<dl>
<dt>Boba Name</dt><dd>{name}</dd>
<dt>City</dt><dd>{city}</dd>
<dt>Rate</dt><dd>{rating}</dd>
<dt>Pikachu</dt><dd>{Number of Pikachu}</dd>
</dl>
"""


boba_locations = top_boba_df[["lat", "long"]]
pika_locations = top_100_pika_df[["lat", "long"]]
parks_locations = top_1000_parks_df[["Latitude", "Longitude"]]
restaurants_locations = resaurant_df[["latitude","longitude"]]
```

```python

fig = gmaps.figure()
boba_symbols=gmaps.symbol_layer(boba_locations)
pika_symbols=gmaps.symbol_layer(pika_locations)
parks_symbols=gmaps.symbol_layer(parks_locations)
restaurants_symbols = gmaps.symbol_layer(restaurants_locations)

#Boba stores were marked green
boba_symbols = gmaps.symbol_layer(boba_locations, fill_color='green', stroke_color='green')
fig.add_layer(boba_symbols)
#Pikachu places were marked yellow
pika_symbols = gmaps.symbol_layer(pika_locations, fill_color='yellow', stroke_color='yellow')
fig.add_layer(pika_symbols)
#Parks were marked red
parks_symbols = gmaps.symbol_layer(parks_locations, fill_color='red', stroke_color='red')
fig.add_layer(parks_symbols)
#Restaurants were marked blue
restaurants_symbols = gmaps.symbol_layer(restaurants_locations, fill_color='blue', stroke_color='blue')
fig.add_layer(restaurants_symbols)
fig

```
